### Import the dataset
First, we will import the dataset, and explore it's attributes

```python
import pandas as pd
df= pd.read_csv('lung_cancer_dataset.csv')
df.head()
df.info()
```

### Data preprocessing

- Drop the `Patient_Id` column


```python
df= df.drop(['Patient Id'], axis=1)
```

- Encode the target variable `Level` 


```python
df['Level']= df['Level'].map({'Low': 0, 'Medium': 1, 'High': 2})
```

- Check for any missing values

```python
df.isnull().sum()
```

No null values found.

After the prepocessing, here is what the dataset loks like:

```python
df.head()
```

### Train-Test split
Split the dataset into features(x) and target (y), and then into training and testing sets. We're diving the dataset into 80:20 ratio for training and testing respectively.

```python
from sklearn.model_selection import train_test_split

X= df.drop(['Level'], axis=1)
y= df['Level']

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=42)
```

### Model Implementation
As per our reference to the research paper, we're going to implement 3 models, namely **SVM**, **Decision Tree**, and **Random Forest**.


- **Support Vector Machine**


```python
from sklearn.svm import SVC

svm= SVC()
svm.fit(X_train, y_train)
svm_predictions= svm.predict(X_test)
```

- **Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier

dt= DecisionTreeClassifier(max_depth=3, random_state=42)
dt.fit(X_train, y_train)
dt_predictions= dt.predict(X_test)
```

- **Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier

rf= RandomForestClassifier(max_depth=5, random_state=42)
rf.fit(X_train, y_train)

rf_predictions= rf.predict(X_test)
```

### Model Performance Evaluation
We're going to be evaluating our models on 3 metrics: **precision**, **recall** and **F1 Score**

First, we're going to generate confusion matrices for each model to visualize the outputs 

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

labels = ['Low Risk', 'Medium Risk', 'High Risk']

svm_cm = confusion_matrix(y_test, svm_predictions)

# Plot the confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(svm_cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.title('SVM Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
rf_cm= confusion_matrix(y_test, rf_predictions)

plt.figure(figsize=(6, 5))
sns.heatmap(rf_cm, annot=True, fmt='d', cmap='Oranges', xticklabels=labels, yticklabels=labels)
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
dt_cm= confusion_matrix(y_test, dt_predictions)

plt.figure(figsize=(6, 5))
sns.heatmap(dt_cm, annot=True, fmt='d', cmap='Greens', xticklabels=labels, yticklabels=labels)
plt.title('Decision Tree Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

We can extract the precision, recall and f1 scores using `classification_report` provided by sklearn

```python
from sklearn.metrics import classification_report

svm_report= classification_report(y_test, svm_predictions, output_dict=True)
rf_report= classification_report(y_test, rf_predictions, output_dict=True)
dt_report= classification_report(y_test, dt_predictions, output_dict=True)
```

Convert it to Dataframes for easy plotting

```python
svm_df = pd.DataFrame(svm_report).transpose()
dt_df = pd.DataFrame(dt_report).transpose()
rf_df = pd.DataFrame(rf_report).transpose()
```

Plot the values

```python
import numpy as np

metrics = ['precision', 'recall', 'f1-score']
categories = ['0', '1', '2']  

svm_values = [svm_df.loc[category, metrics] for category in categories]
dt_values = [dt_df.loc[category, metrics] for category in categories]
rf_values = [rf_df.loc[category, metrics] for category in categories]

x = np.arange(len(metrics))  
width = 0.2 

fig, ax = plt.subplots(3, figsize=(10, 12))

for i, category in enumerate(categories):
    ax[i].bar(x - width, svm_values[i], width, label='SVM', color='blue')
    ax[i].bar(x, dt_values[i], width, label='Decision Tree', color='green')
    ax[i].bar(x + width, rf_values[i], width, label='Random Forest', color='orange')
    
    ax[i].set_title(f'Class {category} (Lung Cancer Level)')
    ax[i].set_xticks(x)
    ax[i].set_xticklabels(metrics)
    ax[i].legend()

plt.tight_layout()
plt.show()
```

Finally, Plot the accuracy scores

```python
from sklearn.metrics import accuracy_score


models = ['SVM', 'Decision Tree', 'Random Forest']
accuracies = [accuracy_score(y_test, svm_predictions), accuracy_score(y_test, dt_predictions), accuracy_score(y_test, rf_predictions)]
plt.bar(models, accuracies)
plt.title('Model Accuracy Comparison')
plt.show()
```

We can see, the accuracy of our models is very high, which may indicate overfitting. We can plot the learning curves for each model to observe how performance changes as the size of the training set increases. A model that overfits will show high accuracy on the training set but low accuracy on the validation set.

For SVM:

```python
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(svm, X, y, cv=5, n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 5))

svm_train_scores_mean = np.mean(train_scores, axis=1)
svm_test_scores_mean = np.mean(test_scores, axis=1)

plt.plot(train_sizes, svm_train_scores_mean, label='Training Score')
plt.plot(train_sizes, svm_test_scores_mean, label='Cross-Validation Score')

plt.title('Learning Curve for SVM')
plt.xlabel('Training Examples')
plt.ylabel('Score')
plt.legend()
plt.show()
```

For Decision Tree:

```python
train_sizes, dt_train_scores, dt_test_scores = learning_curve(dt, X, y, cv=5, n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 5))

dt_train_scores_mean = np.mean(dt_train_scores, axis=1)
dt_test_scores_mean = np.mean(dt_test_scores, axis=1)

plt.plot(train_sizes, dt_train_scores_mean, label='Training Score', color='green')
plt.plot(train_sizes, dt_test_scores_mean , label='Cross-Validation Score', color='orange')

plt.title('Learning Curve for Decision Tree')
plt.xlabel('Training Examples')
plt.ylabel('Score')
plt.legend()
plt.show()
```

For Random Forest:

```python
train_sizes, rf_train_scores, rf_test_scores = learning_curve(rf, X, y, cv=5, n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 5))

rf_train_scores_mean = np.mean(rf_train_scores, axis=1)
rf_test_scores_mean = np.mean(rf_test_scores, axis=1)

plt.plot(train_sizes, rf_train_scores_mean, label='Training Score', color='blue')
plt.plot(train_sizes, rf_test_scores_mean, label='Cross-Validation Score', color='red')

plt.title('Learning Curve for Random Forest')
plt.xlabel('Training Examples')
plt.ylabel('Score')
plt.legend()
plt.show()
```

We can see, the learning curve for SVM and Decision tree show reasonable deviation of the training and cross validation score. This suggests that the models are learning, but there's some generalization error, which is expected in any model. 

However, the learning curve for random forest coincides. This may indicate that the random forest model may be slightly overfitted for the data, but high capacity models like Random Forests can often generalize well without overfitting. The coinciding curves might just indicate that the model has found a good fit for both the training and validation sets.

We can generate a reasonable expection that the models are not overfitted in this case.
